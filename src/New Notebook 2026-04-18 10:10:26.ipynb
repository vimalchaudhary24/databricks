```python
df = spark.read.format("excel") \
    .option("header", "True") \
    .option("inferSchema", "true") \
    .load("s3://s3-databrick-final/src/statement_excel_latest.xlsx")

display(df)
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("edl_dev.databricks_ci_cd.account_statement")
```
